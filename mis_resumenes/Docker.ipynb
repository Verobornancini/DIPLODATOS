## Pasos para utilizar docker

### **1)** Instalar la última versión de Docker disponible:   
https://docs.docker.com/engine/install/
* En la consola podés chequear la versión que tenés:     
<span style="color:red">$ docker --version </span>

### **2)** Creación del **Dockerfile**:   
* Cada contenedor tiene su propio dockerfile.
* En este archivo se setea el sistema operativo a usarse en el contenedor.
* Es una lista de sentencias que definirán el entorno de trabajo.
* Levantará todo lo que esté guardado en la misma carpeta que el dockerfile.

**Ejemplo:**

* **FROM**: Indica la imagen de base a partir de la cual contruiremos la imagen del dockerfile.
* **WORDIR**: Setea la dirección del directorio de trabajo.
* **MAINTAINER**: Nombre de quien construye la imagen.
* **ENV HOME**: Establece el directorio HOME.
* **RUN**: Permite ejecutar las instrucciones: instalar paquetes, etc.
* **ADD**: Añada archivvos al contenedor.
* **EXPOSE**: Indica los puertos por los que se accede a los servicios del contenedor.

### **2)** Creación de la Imagen:   
* A partir del **dockerfile** creamos la imagen utilizando el comando:    
<span style="color:red">$ docker build -t **nombre_imagen** </span>

### **3)** Correr la Imagen y crear el Contenedor:  
* En la consola es posible chequear que se haya construido la **imagen** con el comando:   
<span style="color:red">$ docker images </span>

### **3)** Correr la Imagen y crear el Contenedor:   
* Para correr la imagen utilizamos el comando:    
<span style="color:red">$ docker container run --name **nombre_contenedor** -d -P **nombre_imagen**  </span>

### **4)** Revisión:
* Podemos chequear que nuestra app dentro de nuestro contenedor funcione correctamente con el comando:  
<span style="color:red">$ curl http://localhost:[port]</span>

## COMANDOS BÁSICOS
* **docker build**: Contruye una imagen nueva.
* **docker tag**: Etiqueta una imagen. Se puede usar el flag -t.
* **docker push**: Sube la imagen al repositorio.
* **docker pull**: Descarga la imagen del repositorio.
* **docker run**: Ejecuta el contenedor.
* **docker stop**: Detiene la ejecución del contenedor.
* **docker ps**: Consulta el estado del contenedor.
* **docker ps -a**: Muestra contenedores que ya corrieron y terminaron.
* **docker rm**: Elimina el contenedor.
* **docker images**: Muestra todas las imágenes que tengo creadas.
* **docker start**: Después de crear el contenedor, lo corre.
* **docker container prune**: Limpia los contenedores que ya terminaron o dieron error.
* **docker image prune**: Limpia las imágenes que ya terminaron o dieron error.
* **docker system prune**: Borra todo. Es como hacer todos los prune juntos.
* **docker search imagen**: Busca todas las versiones de las imágenes.
