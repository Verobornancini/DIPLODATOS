## Conceptos importantes   

**Separabilidad Lineal**
* La separabilidad lineal es una propiedad del conjunto de datos con el que estamos trabajando y no del modelo que vamos a entrenar.
* Un conjunto es linealmente separable si puedo encontrar al menos una recta que me separe los dos grupos.

**Clasificadores lineales**    

* Las entradas son vectores de n componenetes (x,y)
* Los clasificadores pueden ser divididos en 2 etapas:    
 **1) Funci√≥n de predicci√≥n** -> arroja un valor num√©rico sobre el cual luego tomo una decisi√≥n dependiendo del umbral de decisi√≥n que haya fijado.   
 **f(x) = W(t) X +W0**    

 **W(t)=** Producto punto de un vector de peso   
 **X=** Vector de datos    
 **W(t) X=** Representa la suma de W1*X1 + W2*X2 + W3*X3......+Wn*Xn      
 **W0=** Par√°metro de sesgo (Bias)   
 
 **2) Funci√≥n de decisi√≥n** -> Seg√∫n el signo del valor num√©rico obtenido en la funci√≥n de predicci√≥n, tomo una decisi√≥n. La decisi√≥n la fijo arbitrariamente a priori.

### 3) Clasificaci√≥n con el Algoritmo del Perceptr√≥n
* Modelo lineal.
* Es un algoritmo desarrollado para clasificar datos en dos categor√≠as (1 √≥ -1), si es que estos son linealmente separables.
* El objetivo es encontrar un hiperplano de separaci√≥n. Un hiper plano es una generalizaci√≥n de una recta a un espacio de n dimensiones. (1 dimension -> recta; 2 dimensiones -> plano; 3 o m√°s dimensiones -> hiperplanos)  
* Este clasificador lineal binario separa el espacio de los datos en 2, esta separacion la hace mediante una frontera. A un lado de la forntera quedan los positivos (1) y al otro lado los negativos (-1).

#### Entrada:
* **Una secuencia de pares de entrenamientos** (x1, y1; x2, Y2....)
* **Un hiperpar√°metro** = Tasa constante de aprendizaje **r** (n√∫mero muy peque√±o y menor a 1)
* Para cada par de datos genero una predicci√≥n seg√∫n mi modelo (+1, -1) y al resultado de esa predicci√≥n lo comparo con lo que tengo en el conjunto de entrenamiento:
 * * Si mi prediccion es correcta, no hago nada. 
 * * Si mi prediccion es incorrecta actualizo el vector de pesos de tal forma que sea igual a la versi√≥n de la iteraci√≥n anterior + la constante r * el producto del vector x,y (W(t+1)= W(t)+r*(x,y))     
* Nuestras incognitas, es decir, lo que buscamos aprender es el valor de los coeficientes de W y W0. Para encontrar una funci√≥n que me separe dos nubes de puntos.
* Si w0=0 -> La recta para separa el hiperplano pasa por el origen. En el algoritmo W0 es al azar o le puedo fijar la semilla a W0.

### Evoluci√≥n del hiperplano cuando se comete un error

* En el gr√°fico anterior, podemos ver que el producto punto del vector (x,y) * W va a ser negativo, ya que W y el vector(x,y) apuntan para lados opuestos. Pero la predicci√≥n dice que debe ser clasificado como positivo -> estoy cometiendo un **error de predicci√≥n**.
* La ecuaci√≥n de actualizaci√≥n me dice que debo sumarle al vector de pesos W la muestra que estoy clasificando mal por un factor de actualizaci√≥n r (W(t+1)= W(t)+r*(x,y)).
* En el gr√°fico y= r   
* El vector W es normal al hiperplano... por eso cuando se mueve el vector W, el hiperplano se desplaza y se actualiza su posici√≥n
* El algoritmo hace esta correcci√≥n por cada par de datos que va analizando   

### Criterio de parada del algoritmo
* Puede ser que decida que una vez que no cometa m√°s errores y ya no se actualice m√°s el hiperplano ahi decida parar de correr el algoritmo..... esto es algo bastante hipot√©tico.
* En la realidad yo nunca s√© si un conjunto es linealmente separable. Por lo tanto en la pr√°ctica el **criterio de parada** suele ser el **n√∫mero de iteraciones** √≥ el n√∫mero de pasadas completas por el conjunto de datos, esa pasada completa se denomina **√©poca**.
* Hay que tener en cuenta que cada vez que yo actualizo el hiperplano debo volver a pasar por todo el cunto otra vez, para chequear que esa actualizaci√≥n no haya generado nuevos errores. Por eso tengo que definir un n√∫mero de √©pocas alto.

### Suponiendo que nuestros datos son linealmente separables...

* Como vemos hay infinitas soluciones al problema de clasificaci√≥n
* Todas las rectas que separen estos puntos son posibles soluciones que puede encontrar el algoritmo del perceptr√≥n.
* La mejor soluci√≥n es aquella que posea el **margen m√°ximo**: es decir que genera el hiperplano m√°s estable ante perturbaciones de la entrada. Es la soluci√≥n que tiene m√°s distancia entre una nube de puntos y la frontera de decisi√≥n. 

### 4) Clasificaci√≥n con el Algoritmo de Vecinos m√°s cercanos (K-Nearest Neighbour)
* Modelo no param√©trico
* Para usar este algoritmo debemos definir:
 * * Una funci√≥n de distancia, que nos de una nocion de cercan√≠
 * * El n√∫mero de elementos que vamos a considerar vecinos cercanos (K). Se suelen elegir n√∫meros impares. Voy a elegir un K tal que el error del conjunto e datos de test sea lo m√°s chico posible.
* Teniendo un set de datos ya clasificados, el modelo lo que hace es tomar el dato nuevo y calcular la distancia del mismo hacia todos los puntos del set de entrenamiento. Luedo ordena en una lista esas distancias y se queada con las K distancias m√°s ceranas que hayamos definido. Entonces a mi nuevo dato finalemnete le asigno la categor√≠a mayoritaria entre los K vecinos m√°s cercanos.

* Para el caso exepcional de K=1 se crea una frontera de decici√≥n entre los dos conjuntos a clasificar (Classificatin boundary) lo que es √∫til para separar conjuntos cuya frontera de separaci√≥n no sea linear.
* Cada fragmento de recta es perpendicual a la linea formada entre la distancia mas corta entre dos de los puntos. Asr√≠ se va formando u entramado que luego dar√° lugar a la forntear de decici√≥n no linear.

* A medida que K aumenta:
 * * La fronterad de desici√≥n se vuelve menos precisa.
 * * Puede aumentar el error en el conjunto de entrenamiento.
* Debo elegir el valor de K √≥ptimo que ser√° el que me d√© el menor valor de error en el conjunto de test.
* Puedo elegir el valor de K utilizando el m√©todo de **Validaci√≥n Cruzada**.
 * * Separo el set de entrenamiento en dos (entrenamiento y validaci√≥n).
 * * Mantengo separado el conjunto de validaci√≥n y mido el error en ese conjunto.

### 5) Clasificaci√≥n con el Algoritmo Regresi√≥n Log√≠stica
* Pertenece a la familia de modelos estad√≠sticos avanzados (lineales generalizados), para conjunto de datos que no poseen distribuci√≥n normal.

* No usa minimos cuadrados como funci√≥n de costo.

# Clasificaci√≥n Multiclase
* Cuando tenemos m√°s de dos categor√≠as para clasificar
* Cada dato puede ser clasificado a k categor√≠as.
* El objetivo es aprender una funcion de predicci√≥n para predecir a que clase pertenece cada dato

#### Abordaje de problemas multiclase
* Una forma es descomponer el modelo multiclase, en muchos modelos de clasificaci√≥n binaria
* Las formas m√°s comunes de descomponer un modelo multiclase son:
 * * Uno vs. todos
 * * Todos vs. todos

#### Uno vs todos
* Asumimos que cada clase puede ser separa de las dem√°s clases
* Entrenamos tantos modelos como clases tengamos
* En cada submodelo se toma a una clase como la positiva (+1) y a todo el resto como clase negativa (-1)

**Funci√≥n de predicci√≥n:** 
 * El scrore m√°ximo gana, se lo conoce como **Winner takes all**
 * El m√°ximo quire decir, el que tenga mayor distancia a la frontera de decisi√≥n
 * Sucede cuando un dato puede ser clasificado como positivo para dos o m√°s clases... la forma de definir a que clase pertenecer√° finalmente es eligiendo aquel modelo que brinde el mayor valor de predicci√≥n.

#### Todos vs todos
* Asumimos que cada par de clases que se puede formar a partir del dataset es separable del resto de las clases.
* Para cada par de clases (j, k) se crea un clasificador binario con:
 * * Ejemplos positivos (categor√≠a j)
 * * Ejemplos negativos (categor√≠a k)
* Cada clasificador creado se especializa en separar entre las dos categorias j y k     

**Funci√≥n de predicci√≥n:** Se realiza mediante un esquema de votaci√≥n. Se puede pensar a cada uno de los clasificadores asociados a cada par de clases, como un elemento que genera una votaci√≥n hacia una de sus categor√≠as.
Ejemplo para un dato: 
* El clasificador (verde-azul) asigna al dato como verde
* El clasificador (verde-rojo) asigna al dato como verde
*  El clasificador (rojo-azul) asigna al dato como azul
En total recibi√≥ dos votos para la clase verde y un voto para la clase azul, por lo tanto ser√° considerado de la clase verde.

* En uno contra todos, uso todas las muestars del conjunto de entrenamiento
* En todos contra todos, en cada clasificador solo considero las muestras correspondientes a las clases pertenecientes a ese par. (cada clasificador es entrenado con un n√∫mero m√°s chico de muestras)     
 **Performance:**
* A nivel de performance no hay mucha diferencia para el entrenamiento (Si tengo 10 clases, aprendo 10 clasificadores)
* La performance a nivel de test a la hora de evaluar una muestra nueva si hay diferencia. en todos vs todos devo evaluar mas clasificadores por lo que puede relentecer un poco(Si tengo 10 clases, aprendo 10*9/2= 45 clasificadores)

**HIPER PAR√ÅMETROS**
* REGRESION POLINOMIAL
 * * 1 Hiperpar√°metro en la funci√≥n de predicci√≥n -> Orden del polinomio.
 * * 1 Hiperpar√°metro en la funci√≥n de costo (error) -> Lamda de regularizaci√≥n (learning rate= r)

# VALIDACI√ìN DEL MODELO ELEGIDO
* Datos se divide en 2: Conjunto de Test y Conjunto de Train
* Conjunto de Train se divide en 2: Train propiamente dicho y Validaci√≥n

* Voy a entrenar tantos modelos, como valores de hiperpar√°metros quiera evaluar
* Cada modelo que entreno lo voy evaluando en el mismo conjunto de validaci√≥n
* Elegir√© luego el modelo con mejor valor de performance
* Paso a unir nuevamente el conjunto de Train propiamente dicho con el de validaci√≥n Y lo corro con el modelo seleccionado
* Ese modelo final es el modelo aprendido con que evaluar√© el conjunto de Test y generar√© un valor de performance final.

* Lo que evaluo como m√©trica de performance es un √∫nico n√∫mero obtenido luego de correr el modelo en el set de test.... al ser un √∫nico valor no tengo como asegurar la estabilidad de la predicci√≥n
* No puedo saber la dispersion o la desviaci√≥n estandar del valor de performance, ya que e sun √∫nico valor
* Para poder obtener estimaciones de estabilidad y robiustez de los modelos hay dos formas:

**1) Remuestreo al azar:** mantengo siempre la proporcion entre conjunto de test y train pero remezclo el data set y tomo de nuevo los conjuntos... asi puedo obtener varios conjuntos de un mismo dataset

* Me permite reportar la media o desviaci√≥n estandar, o int√©rvalo de confianza de la m√©trica que eleg√≠ como representante de la performance del modelo.

**2) Muestreo estratificado:** Cuando el conjunto de entrenamiento es chico, me puede interesar que los subconjuntos que se generen respeten la proporcion de categor√≠as o clases originales. Para eso genero particiones y garantizo que se mantiene la proporcion de clases. Util para problemas multiclase y problemas con datos muy desbalanceados. 

**3) Validaci√≥n Cruzada:** Es el m√©todo m√°s usado. Parto los datos en train y test. y en el conjunto de Train lo divido en n partes iguales (S1, S2, S3....). Lo que hago es tomar alternativamente cada una de las partes como conjunto de validaci√≥n, y al resto como conjunto de entrenamiento. Asi obtengo distintos modelos con distintos valores de performance.
Asi puedo obtener un valor de performance medio en el proceso de selecci√≥n de modelos.
Finalmente uno todos los subsets y corro el modelo elegido, que ser√° el qu√© comparar√© con el conjunto de test.

# M√©tricas de clasificaci√≥n
### M√©tricas y medidas de performance en problemas de clasificaci√≥n

**1)** Antes de comenzar a trabajar debo dividir mi base de datos en dos conjuntos: Train y Test.    
**2)** Trabajaremos optimizando el modelo sobre el conjunto de entrenamiento (Train).     
**3)** Una vez entrenado el modelo hacemos una prueba final sobre el conjunto de test, para tener una estimaci√≥n de la performance (metricas que reflejan cuan bien se esta comportando nuestro modelo).      
**4)** Algunas medidas de performance son:

* Matriz de confusi√≥n
* Accuracy
* Precision
* Recall (Tasa de verdaderos positivos)
* F1 (Tasa de falsos positivos)

**5)** Para obtener una estimacion no sesgada debemos guardar el conjunto de test y no usarlo para nada durante el entrenamiento. Solo lo usamos al final para para computar el valor de performance.

**Matriz de confusi√≥n:** 
* Es muy usada en **problemas multiclase**. En las filas se representan las clases actuales y en las columnas las predicciones.
* En un sistema perfecto todas las predicciones se deber√≠an ver en la diagonal principal, ya que todas las clases actuales estr√≠an predichas correctamente.
* La informaci√≥n √∫til que puedo obtener es cuando veo desviaciones de ese comportamiento (cuadros grises en la matriz). De este modo puedo guiar la toma de decisiones para mejorar mi modelo (por ejemplo, quizas deba unir dos categorias en una, o puedo buscar mas ejemplos, redefinir las variables... etc).
* Siempre empezamos por tratar de mejorar las categor√≠as que tienen las tazas m√°s grandes de de error.

* Tambi√©n es util en **problemas binarios**, es decir de dos soluciones:

* Si tengo un **problema con categor√≠as muy desvalanceadas**, la medida de **accuracy** no es muy adecuada, Por eso hay que saber la metrica a elegir
* Por lo tanto hay otras opciones de medidas de precisi√≥n que se pueden utilizar:

* Otra forma de evaluar la performance del modelo, es  mediante **la Curva ROC**:
* En la curva ROC, elijo un umbral de comparaci√≥n (threshold) y con ese valor evaluo las muestras de Test que dar√° lugar a un valor de Recall y F1. Si evaluo entonces distintos valores de ese humbral, obteng distintos pares de datos de Recall y F1 que me permitir√°n graficar la curva ROC.

# Optimizaci√≥n y Aprendizaje
* Un problema t√≠pico en ML se puede escribir como:
 * * costo de predicci√≥n del par (xi, yi)
 * * regularizaci√≥n para controlar el overfitting

### Optimizaci√≥n Sin restricciones:
* Para una funci√≥n R->R (un n√∫mero escalar me d√° como resultado un n√∫mero escalar) 
* Si quiero encontrar el m√≠nimo de una funci√≥n, la forma es calcular la derivada de esa funci√≥n respecto de la variable en estudio.
* El punto en el que la derivada de una funci√≥n es =0 se llama **Punto estacionario**. Hay tres tipos de puntos estacionarios:

* En todos estos puntos, la derivada (pensada como la tangente de la curva) es =0
* Calculando la derivada segunda, puedo saber si ese punto es un m√≠nimo, un m√°ximo o un punto de inflecci√≥n.
* Esta idea de los puntos estacionarios se puede extender a funciones n dimensionales, Rn -> R (un vector me d√° como resultado un n√∫mero escalar) con el nombre de **Vector Gradiente**
* La funcion f tiene un punto estacionario en X0 (f(X0)) del vector cuando se d√° la condici√≥n que X0 valga 0
* X0 va a ser un punto estacionario, si yo calculo el vector gradiente en esa coordenada y el valor del vector es cero
* En el caso de funciones Rn, el concepto de derivada segunda, se extiende al concepto de **Matriz hessiana**, que es una derivada donde estan todos los pares de derivadas en distintos √≥rdenes (respecto a x1, x2, x3 ,etc)
* **Vector gradiente:** calculado mediante la funci√≥n de costo (W). La direcci√≥n de este vector, me da informaci√≥n del sentido en el que crece la funci√≥n.
* El vector gradiente siempre es normal a las lineas de nivel de la funci√≥n. Las l√≠neas de nivel son aquellas en las cuales el valor de la funci√≥n es constante (vale lo mismo).

* **M√©todos iterativos de descenso:** A partir de un W avanzo en una direcci√≥n decreciente, Es decir que el valor de la funci√≥n valuada en ese punto es menor que el valuado en el punto anterior. Para dar cada paso descendente, me muevo en el sentido contrario al que me indique el vector gradiente.

## Algoritmos de Descenso de gradiente
* delta W(t): direccion y magnitud del moviento del paso desdentente
* delta W(t): C * gradiente de la funci√≥n de costo
 * * C= coeficiente: indica la magnitud del paso
