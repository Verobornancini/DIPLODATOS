# Algoritmos de Clasificaci√≥n
## Naive Bayes
* Es un clasificador basado en el teorema de Bayes, con una asunci√≥n ‚Äúnaive‚Äù sobre los datos.
* Es muy sencillo de programar y entender.
* Sirve mucho como baseline y, aunque simplista, puede tener resultados que sobrepasan a algoritmos mucho m√°s complejos.
* Es r√°pido de entrenar y funciona con datos de mucha dimensionalidad (e.g. es muy √∫til a la hora de clasificar documentos).
* El algoritmo de ‚ÄúNaive Bayes‚Äù  est√° fuertemente ligado al teorema de Bayes.
* El teorema de Bayes establece que se puede encontrar la probabilidad de A (e.g. una clase objetivo) dada la ocurrencia de B (e.g. un conjunto de features). 
* Es decir, B es la evidencia (features) y A es la hip√≥tesis (clases).   
#### El Teorema de Bayes establece:

* Bajo el teorema de Bayes, la principal asunci√≥n es que los atributos son independientes entre s√≠. 
* La presencia de un feature no afecta a los otros. Esta asunci√≥n es ‚Äúnaive‚Äù, por eso el nombre del algoritmo.
* Una segunda asunci√≥n, es que todos los atributos tienen el mismo efecto en la salida del algoritmo.

### SVM: Outliers
* La mayor√≠a de los casos, los datos no son linealmente separables. Tiene el mismo problema del perceptron, por eso a√±aden un par√°metro C.
* En algunos casos, existen outliers.
* Hay un par√°metro que define qu√© tan tolerante puede ser SVM sobre la clasificaci√≥n incorrecta de datos.
* El ‚Äúpar√°metro C‚Äù, (es un hiperpar√°metro que debemos fijar) define un tradeoff entre clasificar mejor los datos de entrenamiento y tener una mejor ‚Äúseparaci√≥n‚Äù (un margen m√°s amplio).


## Datos no linealmente separables
* SVM es una t√©cnica para separar los datos mediante un hiperplano.
* Si los datos no son linealmente separables, dicho hiperplano no existe.
* **Soluci√≥n:** Proyectar los datos a una dimensi√≥n donde s√≠ sean linealmente separables. Agrego una nueva dimensi√≥n con las siguientes ecuaciones para cada dimensi√≥n. Ahora si puedo separar los datos con un hiperplano.

* Kernel de este ejemplo: **K (xi, xj)= (xi, xj)2**

## Kernel
* Un Kernel es una funci√≥n que proyecta los datos de una dimension a otras dimensiones, para poder encontrar planos de separaci√≥n.
* El kernel es un hiperpar√°metro a definir.
* El kernel lineal no hace cambio de dimensiones, seria lo mismo que una recta en un plano.
* La manera en que el algoritmo de SVM realiza la proyecci√≥n es mediante el uso de kernels.
* Las funciones de kernel toma dos puntos del espacio original, y devuelve el producto punto en el espacio proyectado. Este producto punto es lo que la funci√≥n de SVM necesita para calcular el costo.

# √Årboles de decisi√≥n
* Aprende a diferenciar los datos en base a reglas de decisi√≥n.
* Los nodos del √°rbol representan las reglas. Las hojas asignan la clase o el valor.
* El √°rbol se particiona recursivamente. Para obtener el resultado, simplemente se siguen los nodos de decisi√≥n de acuerdo a los datos y se asigna la clase final.
* Es un algoritmo de ‚Äúcaja blanca‚Äù, ya que puede visualizarse f√°cilmente e interpretarse por los humanos (a diferencia de un algoritmo de ‚Äúcaja negra‚Äù como son las redes neuronales).
* Son buenos con datos de mucha dimensionalidad (high dimensional data)

**El algoritmo:**
* Selecciona el mejor atributo de acuerdo a alguna m√©trica (e.g. ganancia de informaci√≥n).
* Hacer un nodo de decisi√≥n con ese atributo, que particione los datos en subconjuntos de menor tama√±o.
* Repetir recursivamente el procedimiento para cada nodo hijo.
* El algoritmo se detiene si:
* * Todos los ejemplos del subconjunto son de la misma clase.
* * Todos los elementos del subconjunto son constantes con respecto al atributo/s de inter√©s del nodo actual.

**M√©todos de decisi√≥n:**
* Definen la forma de particionar el conjunto de datos (es una heur√≠stica).
* Buscan rankear cada atributo de acuerdo a ciertos par√°metros.
* Los criterios m√°s populares son:
* * **Information Gain:** calcula la diferencia entre la entrop√≠a del conjunto antes de dividirse y la entrop√≠a media luego de dividir el conjunto de datos, para cada atributo.
* * **Gini Index:** Es una m√©trica que mide cu√°nto un elemento del subconjunto elegido al azar ser√≠a identificado incorrectamente. Busca los atributos que tienen menor √≠ndice de Gini.
* * **Mean Squared Error:** Este m√©todo sirve para casos de regresi√≥n (en lugar de clasificaci√≥n). Busca minimizar el error cuadr√°tico medio en los nodos hijos a la hora de particionar los datos.

# Ensemble Learning
* Se basan en la idea de que el trabajo en conjunto deber√≠a dar mejores resultados.
* De tal forma, un conjunto de modelos, al combinarse, deber√≠an tener mejor performance.
* Consideremos el caso de tres modelos: M1, M2 y M3:
* Cuando las predicciones son iguales, es sencillo definir c√≥mo predecir; qu√© hacemos en el caso en que difieran?
* * Le creemos al mejor de los modelos?
* * Votamos por mayor√≠a?
* Un modelo "ensemble" se constituye como un conjunto de diferentes modelos.
* Habitualmente, un modelo "ensemble" es m√°s preciso que los modelos que lo constituyen. Intuitivamente, esto se debe a que "dos aprenden mejor que uno".     
**Aproximaci√≥n de justificaci√≥n**
* Asumimos que tenemos un modelo M, formado por n modelos: M1, M2,... Mn.
* Cuando un modelo recibe un dato x, el modelo predice M(x) a partir de las predicciones Mi(x), a partir de votaci√≥n (clase m√°s votada).
C√≥mo determinamos cu√°n bien funciona el modelo?
* Para responder, consideramos:
* * Todos los clasificadores M1, M2,... Mn son igualmente precisos (precisi√≥n p)
* * Los errores en la clasificaci√≥n hecha por cada clasificador son independientes: P(Mj err√≥neo | Mk err√≥neo) = P(Mj err√≥neo)
* Para hacer el ejemplo m√°s concreto, consideremos:      
p = 0.8    
n = 5    
Entonces, P(la predicci√≥n de M es correcta) = P(al menos 3 de los 5 Mi predicen correctamente)

# Modelos de Ensemble Learning
## 1) Bagging 
* Es un m√©todo para hacer aprendizaje por "ensemble".
* Si usamos el mismo modelo sobre los mismos datos, obtendremos los mismos resultados (salvo inicializaciones aleatorias). Entonces, de cierta forma, debemos introducir variaciones en los datos:
* Si D es el dataset inicial; repetimos k veces lo siguiente:
* * Generar Di a partir de las entradas de D, seleccionando aleatoriamente y con reposici√≥n |D| instancias de D.
* * Entrenamos el modelo Mi a partir de Di.
* * Nuestro modelo M selecciona las predicciones m√°s frecuentes de {Mi}i.

* Cada re arreglo de datos, genera un arbol diferente... luego por un sistema de votaci√≥n se selecciona el arbol que m√°s veces se logr√≥ construir.

* **Bagging para √°rboles de decisi√≥n** (es un modelo que utiliza otros modelos, no siepre tienen que ser arboles de decisi√≥n)
* Bagging generalmente funciona bien para algoritmos "inestables":
* Un algoritmo es inestables si peque√±as variaciones en el dataset pueden generar modelos muy diferentes.
* Resulta que los √°rboles de decisi√≥n son inestables.
* Si bien lo anterior incentiva al uso de bagging para √°rboles de decisi√≥n, tenemos una desventaja importante: entrenar k √°rboles es k veces m√°s caro que entrenar uno solo.

## 2) Random Forests
* Las Random Forests son una modificaci√≥n a Bagging para √Årboles de Decisi√≥n. Para evitar la sobrecarga, se simplifican los modelos:
* Como vimos antes, cuando se crean los √°rboles (cuando se entrenan), todas las features son consideradas o evaluadas al crear cada nodo.
* Para los Random Forests, en cada nodo se consideran s√≥lo M atributos elegidos aleatoriamente (par√°metro max_features en sklearn).
* Usualmente, se toma: M= raiz cuadrada del n√∫mero de atributos

* El modelo se puede resumir en los siguientes pasos:
* Repetir k veces:
* * Construir un dataset a partir del original, como se hace con bagging.
* * Construir un √°rbol de decisi√≥n:
* * * Para cada nodo del √°rbol, seleccionar M variables y construir el nodo "√≥ptimo" entre esas features.
* * * Repetir hasta que el √°rbol est√© completamente construido (no se hace pruning).


# Boosting
* M√©todo para hacer aprendizaje por "ensemble".
* Utilizamos distintos modelos para generar votos. la diferencia con lo anterior es la importancia que le damos a cada voto.
* Para el ejemplo de los √°rboles de decisi√≥n, vimos que con bagging, entrenamos grandes √°rboles sobre versiones del material de entrenamiento generados con reposici√≥n; y luego votamos por mayor√≠a.
* Boosting pretende darle mayor importancia a los modelos que mejor performance tienen sobre los datos. As√≠, podemos decir que, con boosting, se entrenan √°rboles sobre versiones "pesadas" de los datos de entrenamiento. Y luego se clasifica por mayor√≠a pero pesando los modelos.
* Cada modelo, de cierta forma, define qu√© features considerar√° el siguiente modelo.
* Se seleccionan datos como en bagging (bootstrapping), pero en este caso, los datos son pesados con cierto criterio. De hecho, algunos registros ser√°n usados m√°s frecuentemente.
* El proceso es el siguiente:
* * Dado un modelo, determinar qu√© registros son m√°s "err√≥neos" y darles mayor pesos para los modelos siguientes).
* * Dado un modelo, determinar su performance para darle mayor peso a los "mejores" modelos.

# Voting
* Clasificador compuesto de varios clasificadores.
* Cada clasificador que lo compone se entrena sobre el conjunto de datos.
* El clasificador de ‚Äúvoting‚Äù final simplemente elige la clase que tuvo ‚Äúm√°s votos‚Äù de parte de los clasificadores que lo componen.
* La votaci√≥n puede ser ‚Äúhard‚Äù (simplemente se cuenta la cantidad de votos para una clase) o ‚Äúsoft‚Äù (se usa la probabilidad).
