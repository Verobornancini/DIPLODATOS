# REDES NEURONALES

* Funcion sigmoide de una regresion logística, como una función de activación de la neurona.
* W= Pesos * X
* X= valor de entrada y qye dsp viene de la neurona anterior (alfa)
* La funcion de activacion tiene que apagar y prender las neuronas para ir avanzando en las capas

* Cada neurona recibe una variable de entrada
* Los pesos de las neuronas se inicializan de manera aletoria en valores distintos.
* El valor que ingresa en la capa de entrada se multiplica por el valor de coste de esa capa.
* La capa de salida, la define mi problema
* Las capas intermedias, se denominan como capas ocultas. permiten que la red neuronal encuentre valores distintos y genere mejor interconeccion entre las variables de entrada.
* Si mi funcion de activación es la ogística, cada capa recibe como imput un valor entre 0 y 1
* Las capas siguientes no necesariamente deben tener menos neuronas que la anterior.

## ¿Que hacer con múltiples clases?
* En vez de obtener un valor de probabilidad de que ocurra la clase positiva... obtenemos un vector de valores de probabilidad que corresponde con cada una de las clases.
* La función de costo cambia
* La idea es tener un vector final en las capas de salida

# Redes neuronales profundas (Deep Learning)
* EJEMPLO CON IMÁGENES:
* Las primeras capas analizan rasgos mas generales, lineas, texturas etc.
* Las capas medias ya detectan formaciones, como ojos, bocas
* Las capas aun mas profundas detectan caras completas
* Se usan modelos preentrenados, y nosotros modificamos las últimas capas y las especializo para clasificar mi problema particular.

## Redes Neuronales: sesgo y varianza
* Underfitting (high bias):
 * * Ampliar la red
 * * Cambiar la arquitectura de la red
* Overfitting (high variance):
 * * Agregar más datos
 * * Regularización
 * * Cambiar la arquitectura de la red

* Dropout hace que se apaguen ciertas neuronas para evitar que la red se memorice los resultados

# Sistemas de Recomendación
* El objetivo es guiar a los usuarios a tomar decisiones (ver una peli, comprar un producto, etc.)

## Tipos de Sistemas de Recomendación
* **Basados en contenido:** tratan el problema de manera específica para cada usuario y "aprenden" una clasificación de lo que a un usuario le gusta basado en las características (contenido) de un ítem. El problema es que muchas veces el usuario no sabe lo que quiere y puede ranquear mas sus intereses.
* **Filtros colaborativos:** se basan en la idea que los usuarios que coinciden en el pasado lo harán en el futuro y que les gustarán ítems similares.
* **Basados en el conocimiento**: el usuario proporciona explícitamente parámetros del producto que quiere.
* **Basados en cuestiones demográficas**
* **Sistemas híbridos**

### Filtros colaborativos
* Combina usuarios con intereses similares
 * * Se podría requerir de muchos usuarios para que las posibilidades de encontrar un par con intereses en común.
 * * Tiene que existir un método sencillo para reflejar los intereses de los usuarios.
 * * Se necesita de un algoritmo eficiente para combinar a los usuarios con gustos similares.
* Como funcionan ?
 * * **INPUT:** Usuarios generan puntuaciones a un conjunto de ítems (implícita o explícitamente): matriz de puntuaciones usuario-ítem 
 * * **OUTPUT:** Predicción (numérica) indicando cuánto a un usuario le gusta un ítem. Una lista con N ítems recomendados por usuario

### Filtros colaborativos: Vecindarios basado en usuarios
* Dado un usuario (Alice) y un ítem que todavía no se ha asignado un puntaje, se debe:
 * * Encontrar el conjunto de usuarios que más se parecen a Alice (usuarios a los que les gustan ítems similares) y han puntuado el objeto
 * * Usar sus puntuaciones para predecir si a Alice le gustará el ítem
 * * Aplicar este proceso sobre todos los objetos que Alice no ha puntuado y recomendar los que tienen mayor puntuación

* Utilizo la media, para evitar el sesgo de votacion de los usuarios.
* Comparo la similitud de Alice respecto de cada uno de los demás usuarios, teniendo en cuenta el coeficiente de pearson (1 a -1)

* Para evitar el sesgo se resta la media
* Se suma la similitud de los items multiplicada por el usuario
* Elijo clasificar por item o por usuario, según cual de mis dos grupos de atos es m´s estable

### Preprocesamiento
* Requerimientos de memoria: si consideramos n ítems, tendremos n2 similitudes. 
* En la práctica, la matriz es dispersa (muchos pares de ítems que no tienen similitud). 
* Se suelen usar reducciones fijando un umbral mínimo de "co-ratings": se eliminan ítems que tienen pocas puntuaciones comunes, usando al menos n' usuarios.
* Se puede también limitar el tamaño del vecindario (N).
