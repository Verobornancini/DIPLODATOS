# <span style="color:deeppink">REST (REpresentational State Transfer)</span>
Arquitectura de software que define restricciones para servicios web y
garantiza interoperabilidad entre sistemas en Internet.
Si la API de un servicio web cumple con esas restricciones diremos que es
una RESTful API.

### <span style="color:deeppink">ESTRUCTURA DE LA DIRECCIÓN DE UN SERVICIO WEB</span>
* **URI** - Identificador de Recursos Uniforme
* **URL** - Locación de Recursos Uniforme
* **URN** - Nombre de Recursos Uniforme

## Operaciones
* **POST** - creación de recursos
* **GET** - obtener un recurso
* **PUT** - reemplazar un recurso
* **PATCH** - actualizar parcialmente un recurso
* **DELETE** - remover un recurso

**FORMATO DE USO:**  
GET /users/:id   
POST /users/:id   
DELETE /users/:id   
PUT /users/:id   
GET /users   
GET/users?name=“smith”   

**Especificación OpenAPI:** Fue creada en 2010. Estándar independiente del lenguaje para definir y describir APIs REST.

**Swagger:** Editor - permite diseñar las APIs en el navegador.

**GraphQL:** Lenguaje para consultar y manipular datos de APIs.

--------------------------------------------------------------
# Protocolos de comunicación
--------------------------------------------------------------

## <span style="color:purple">Polling</span>

**Short Polling:** El cliente envía requests “constantemente” al servidor consultando si hay datos de interés. Genera tráfico muchas veces innecesario.

**Long Polling:** El cliente envía request al servidor que mantiene abierta la conexión hasta que tiene una respuesta para el cliente. Genera un consumo de recursos más alto en el servidor.

**Periodic Polling:** El cliente envía requests al servidor a intervalos regulares. Equilibrio entre el uso de recursos del servidor y el tráfico entre ambas partes.

## <span style="color:purple">Streaming</span>   
El servidor envía una respuesta de longitud indefinida. Un intermediario entre el cliente y el servidor (proxy) puede interrumpir la conexión.No hay garantías de que el contenido llegue a tiempo.

## <span style="color:purple">Webhooks</span>   
Registro de una URL de “callback” entre servidores

# <span style="color:purple">RSS</span>

* RDF Site Summary / Really SimpleSyndication. 
* Mecanismo de Publicación/Subscripción para recibir el contenido actualizado de un sitio.
* XML estándar para redifusión de contenidos.

# <span style="color:purple">Feed Parser</span>

Permite descargar contenidos de un **feed RSS** -> feedparser.parse(rss_url)   
https://pythonhosted.org/feedparser/   

```python
# EJEMPLO:
# Subscripción para leer todas las entradas que hay en la página de la Voz del Interior

import feedparser
lavoz_data = feedparser.parse("https://www.lavoz.com.ar/rss/tecnologia.xml")
```

```python
# Por cada entrada vamos imprimiendo todos los títulos

for item in lavoz_data['entries']:
    print("-", item['title']) 
```

# <span style="color:purple">WebSocket</span>
* Protocolo de comunicación full-duplex en una única conexión TCP.
* Menos sobrecarga y más flexibilidad que HTTP Polling.
* Transferencia de datos “en tiempo real”.
* Puertos estándar HTTP (80, 443).
* Contenido binario no HTTP.
* Estandarizado en 2011.
