**Ejercicio 4.1:** Canciones de Iron Maiden

```python
SELECT * FROM tracks 
INNER JOIN albums on tracks.AlbumId = albums.AlbumId
INNER JOIN artists on albums.ArtistId = artists.ArtistId
WHERE artists.name = 'Iron Maiden'
```

**Ejercicio 4.2:** Discos con más de 25 canciones

```python
SELECT al.title, COUNT(1) cantidad_canciones FROM tracks tr, albums al
WHERE al.albumid = tr.albumid
Group By tr.AlbumId
HAVING COUNT(1) > 25
```

**Ejercicio 4.3:** Las 10 canciones más populares

```python
SELECT tr.name, count(1) cant_esc
FROM playlist_track pt
JOIN tracks tr ON pt.TrackId = tr.TrackId
GROUP BY tr.name
ORDER BY 2 desc
LIMIT 10;
```
